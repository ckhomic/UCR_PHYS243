```python
def Fib_rec (n=0):
    if n==0:
        return 1
    elif n==1:
        return 1
    else:
        return Fib_rec(n-1)+Fib_rec(n-2)
```

```python
def Fib_for(n=0):
    a=0
    b=1
    for i in range (0,n):
        temp=a
        a=b
        b=temp+a
    return b
```

Part 2: The for loop runs faster because Recurseive function calls the function itself twice.

Part 3

```python
import time
# if you run this, the current time in (s) will be recorded in x
x=time.time()
```

```python
def timer(n,k,f=Fib_rec):
    time_start=time.time()
    for i in range (0,k):
        f(n)
    time_end=time.time()
    average_time=(time_end-time_start)/k
    return average_time
```

Part 4

```python
import matplotlib.pylab as plt
import numpy as np
```

```python
# We input n=20, k=30 as the number of runs
n=20
x=[]
time_rec=[]
time_loop=[]

for i in range (1,n+1):
    x=np.append(x,i)
    time_rec=np.append(time_rec,timer(i,30,Fib_rec))
    time_loop=np.append(time_loop,timer(i,30,Fib_for))
```

```python
#Plot graph
plt.figure(figsize=(8,8))
plt.plot(x,time_rec,c='darkred',label="Recursive Fibonacci")
plt.plot(x,time_loop,c='darkblue',label="Fibonacci For Loop")
plt.xlabel('n',fontsize=15)
plt.ylabel('Average time',fontsize=15)
plt.legend(fontsize=15)
plt.show()
```
